```python
#Location, start date, end date, 
import pandas as pd
import requests
import json
url = "http://eventful.com/oauth/authorize?oauth_token=ZXRjQzTcLvs84Dqh"
states = []

```

```python
states = []
file = "states.csv"
states_df = pd.read_csv(file)
for i in states_df['State']:
    states.append(i.replace(" ","+"))
states
```

```python
event_title = []
city = []
start_time  = []
latitude = []
longitude = []
short_bio = []
going_count = []
state_df = []
for state in states:
    search_url = f"http://api.eventful.com/json/events/search?...&location={state}&date=future&sort_order=popularity&page_size=10&app_key=ZXRjQzTcLvs84Dqh"
    test = requests.get(search_url).json()
    items = test[]
    for item in items[:20]:
        event_title.append(item['title'])
        city.append(item['city_name'])
        state_df.append(item['region_name'])
        start_time.append(item['start_time'][:10])
        latitude.append(item['latitude'])
        longitude.append(item['longitude'])
        if isinstance(item['performers']['performer'],(list,)):      
            short_bio.append(item['performers']['performer'][0]['short_bio'])
        else:
            short_bio.append(item['performers']['performer']['short_bio'])
        going_count.append(item['going_count'])





```

```python
events_df = pd.DataFrame({"Title":event_title,
                          "City": city,
                          "State": state_df,
                          "Start Time": start_time,
                          "Latitude": latitude,
                          "Longitude": longitude,
                          "# Going": going_count})
state_df
```

```python
df = events_df

ustates = df['State'].value_counts()
ustates = pd.DataFrame(ustates)
ustates.reset_index()
ustates
```
